# Data Preprocessing

*Importing the Required Libraries*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sm
```

*Importing the Dataset*

```python
dataset = pd.read_csv('Terrorism.csv', encoding='latin-1')
```

*Displaying the imported Dataset*

```python
dataset.describe()
```

*Displaying the datatype of Dataset*

*Shaping the Dataset*

```python
dataset.shape
```

```python
dataset.info()
```

*Renaming the Required Columns*

```python
dataset.columns
```

```python
dataset = dataset.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country', 'region_txt':'Region', 'provstate':'Province State', 'city':'City', 'attacktype1_txt':'Attack Type', 'targtype1_txt':'Target Type','targsubtype1_txt':'Target SubType', 'gname':'Gang Name', 'weaptype1_txt':'Weapon Type', 'weapdetail':'Weapon Detail'})
dataset = dataset[['Year', 'Month', 'Day', 'Country', 'Region', 'Province State', 'City', 'Attack Type', 'Target Type','Target SubType', 'Gang Name', 'Weapon Type', 'Weapon Detail']]
```

```python
dataset
```

```python
dataset.isnull()
```

*Finding and Fixing the Missing Values*.

*We can find Missing values and fill them using mean or median or most-frequent*

```python
missing_val = dataset.isnull().sum()
missing_val
```

```python
dataset.fillna(dataset.mean(numeric_only=True).round(1), inplace=True)
dataset
```

*Finding the Coorelation between the various columns*

```python
dataset.corr()
```

# Visualizing the Dataset

*Top 10 Countries affected by terrorism attacks*

```python
title_tp = dataset['Country'].value_counts()[:10]
plt.figure(figsize=(10,6))
sm.barplot(x=title_tp.index, y=title_tp.values)
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.title('Top 10 Countries affected by Terror Attacks')
plt.xticks(rotation=45)
plt.show()
```

*Distribution of Terror Attacks across various regions of World*

```python
tl_tp = dataset['Country'].value_counts().index
plt.figure(figsize=(11,11))
title_tp.plot(kind = 'pie')
plt.title('Terror Attacks across various regions')
plt.show()
```

*Most Frequent Types of Terrorist Attacks*

```python
plt.figure(figsize=(12, 6))
sm.countplot(x='Attack Type', data=dataset, order=dataset['Attack Type'].value_counts().index)
plt.title('Most Frequent Attack Types')
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

*Attacks based on Different Months*

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=90)
plt.title('Attacks on different months')
sm.countplot(x=dataset['Month'])
```

*Number of Terrorism Activities per year*

```python
plt.figure(figsize=(12,10))  # Adjust the figure size as needed
plt.title('Types of Weapon Used')
weapon_counts = dataset['Weapon Type'].value_counts()
sm.barplot(x=weapon_counts.values, y=weapon_counts.index, palette='pastel')
plt.xlabel('Count')
plt.ylabel('Weapon Type')
plt.show()
```

*Types of Targets*

```python
plt.figure(figsize=(19,6))
plt.xticks(rotation=45)
plt.title('Target Type')
sm.countplot(x='Target Type', data=dataset, palette='viridis')
plt.xlabel('Target Type')
plt.ylabel('Count')
plt.show()
```

*Coorelation Heat-Map*

```python
plt.figure(figsize=(17,12))
coor_mat = dataset.corr()
sm.heatmap(coor_mat, annot=True, cmap='coolwarm')
plt.title("Coorelation Heatmap")
plt.show()
```
